#### 1. Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

Точность (precision) — показывает, сколько из предсказанных позитивных объектов, оказались действительно позитивными.<br>

Ситуации, когда лучше максимизировать Precision:<br>

1. Предсказание оттока клиента: в случае ограниченности ресурсов компании (например возможностей колл-центра совершать только X звонков), выгодней искать не всех клиентов, которые уйдут в отток, а тех, чья вероятность ухода будет наибольшей (max(Precision) --> 1) и соответственно звонить уже им с целью удержания.
2. Предсказание кредитного риска: в случае если нам необходимо зафиксировать порог невозврата кредита в размере 5% для уменьшения уровня просрочки портфеля, то нам необходимо построить классификатор с Precision = 0.95, при этом нам не столь важна метрика Recall в данной постановке задачи (здесь также можно так варьировать порогом t, чтобы min(Precision) = 0.95, а max(Recall) --> 1, то есть сначала отобрать пороги t при которых Precision >= 0.95, а затем уже ранжировать по max(Recall)).


Полнота (recall) — показывает, сколько от общего числа реальных позитивных объектов, было предсказано, как позитивный класс.<br>

Ситуации, когда лучше максимизировать Recall:<br>
1. Предсказание наличия рака у пациента: в данной постановке задачи лучше направить пациента на дополнительные анализы и проверки, чем пропустить больного, поэтому прежде всего важна максимальная полнота.
2. Поиск мошеннических операций: предположим нам необходимо найти все сомнительные операции клиентов, в таком случае, нам необходимо выбрать модель с максимальной полнотой, так как зачастую, пропуск "сомнительных" операций может привести к риску санкций со стороны регулятора (например приостановка деятельности всей компании) или финансовым потерям и репутационным рискам. В данной задаче лучше дополнительно отправить операции клиента на "ручную" проверку, чем совем пропустить такие операции, что может привести к гораздо большим рискам (например отзыв лицензии для банков).

#### 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Пусть precision = 0.1, recall = 0.9.<br>

Предположим, что $F_{1}$-мера = (precision + recall) / 2 = (0.1 + 0.9) / 2 = 0.5. В таком случаем, $F_{1}$-мера это среднее арифметическое из точности и полноты. При слишком большой разнице между точностью и полнотой (например 0.1 и 0.9) мы усредняем оценку и не учитываем низкое значение точности (0.1) в данном примере.<br>

Предположим, что $F_{1}$-мера = (2 * precision * recall) / (precisiob + recall) = (2 * 0.1 * 0.9) / (0.1 + 0.9) = 0.18. В таком случаем, $F_{1}$-мера это среднее гармоническое между точностью и полнотой. При слишком большой разнице между точностью и полнотой (например 0.1 и 0.9) мы учитываем низкое значение точности (0.1) в данном примере и оценка $F_{1}-мера$ будет более консервативной, чем среднее арифметическое.<br>

Преимущество $F_{1}-меры$ заключается в том, что данная метрика дает более сбалансированную оценку качества модели (алгоритма), чем точность и полнота по отдельности. Она стремится к нулю, если точность или полнота стремятся к нулю, и наоборот, стремится к единице, если точность и полнота стремятся к единице. Если, например, точность очень мала, то увеличение полноты, пусть и в два раза, не сильно меняет значение функционала. <br>

$F_{1}-мера$ сводит к одному числу две других основополагающих метрики: точность и полноту, что является более удобным при сранвнении результата работы нескольких моделей и принятии решений о выборе лучшей. <br>
