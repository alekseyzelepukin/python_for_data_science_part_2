#### 1. В чём различие между зависимыми и независимыми выборками?

Если можно установить гомоморфную пару (то есть, когда одному случаю из выборки X сооветствует один и только один случай из выборки Y и наоборот) для каждого случая в двух выборках (и это основание взаимосвязи является важным для измеряемого на выборках признака), такие выборки называются зависимыми. В случае, если такая взаимосвязь между выборками отсутствует, то эти выборки считаются независимыми.<br>

Иными слова, зависимую выборку (как правило / в большинстве случаев) составляют одни и те же объекты наблюдения, например пациенты до и после приема лекарства. Каждому пациенту из выборки X0 будет соответствовать он же сам из выборки X1, но уже после приема медицинского препарата. Иногда степень зависимости между объектами может быть не такая сильная (строгая), например при изучении мужей и жен / братьев и сестер, но и в этом случае, зависимость между изучаемыми объектами сохраняется (от попадания конкретного мужа в выборку Y0 зависит попадание жены в выборку Y1).<br>

Независимые выборки характеризуются тем, что вероятность отбора наблюдения из одной выборки не зависит от вероятности отбора любого наблюдения из другой выборки. В качестве примера можно привести случайный отбор клиентов компании в тестовую и контрольную группы (выборки) для тестирования эффекта целевого предложения (например нового тарифа или пакета услуг). Каждому клиенту из тестовой группы не будет соответстовать парный объект из контрольной группы. На практике часто встречается, что целевая и контрольные группы отличаются друг от друга размером (например: в тестовую группу попали в результате случайного отбора 95% всех клиентов, а в контрольную в результате того же случайного отбора только 5% всех клиентов, при этом клиенты в каждой из групп независимы друг от друга).<br>

#### 2. Когда применяются параметрические статистические критерии, а когда их непараметрические аналоги?

Параметрические критерии - группа статистических критериев, которые включают в расчет параметры вероятностного распределения признака (средние и дисперсии). Параметрические критерии, применяются при известном законе распредления генеральной совокупности, как правило нормальном или при использовании параметров данной совокупности (среднее, дисперсия). При нормальном распределении генеральной совокупности параметрические критерии обладают большей мощностью по сравнению с непараметрическими. Иными словами, они способны с большей достоверностью отвергать нулевую гипотезу, если последняя неверна. <br>

Непараметрические критерии - группа статистических критериев, которые не включают в расчёт параметры вероятностного распределения и основаны на оперировании частотами или рангами. На практике, большинство изучаемых данных, не подлежат номарльному закону распределния. Применение в таких случаях параметрических критериев может привести к ошибкам в статистических выводах. В таких случаях непараметрические критерии оказываются более мощными, т.е. способными с большей достоверностью отвергать нулевую гипотезу. <br>
