#### 1. Для чего и в каких случаях полезны различные варианты усреднения лейблов целевой переменной: micro, macro, weighted?

Варианты усреднения метрик качества работы модели полезно использовать в задачах мультиклассовой классификации.<br>

1. Macro - это средняя оценка метрики качества без учета веса (размера) каждого класса. Данное усреднение используется для оценивания качества работы классификатора в целом, на всех классах сразу. Полезно использовать, когда важно оценить качество классификатора на данных, в которых присутствуют классы небольшого размера и верное предсказание этих классов критично.<br>
2. Weighted - это средняя оценка метрики качества с учетом веса (размера) каждого класса. Итоговая оценка метрики качества взвешивается на размер каждого класса, полезно использовать при наличии дисбаланса классов в данных. Так как небольшим по размеру классам будет присваиваться мешьший вес в итоговой метрике качества, данное усредниение стоит использовать когда качество работы классификатора на малых классах не является критичным.<br>
3. Micro - это средняя оценка метрики качества, которая не усредняет отдельные оценки метрик качества на разных классах, а суммирует составляющие самой метрики (TP, FP, etc.) и позволяет получить общую агрегированную оценку. Полезно использовать, когда необходимо оценить максимальное качество работы классификатора и при этом качество предсказаний миноритарного класса не критично.<br>

#### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

XGBoost, LightGBM, Catboost - это различные реализации алгоритма градиентного бустинга на решающих деревьях (GBM).<br>

LightGBM использует более "новый" алгоритм Gradient-based One-Side Sampling для нахождения наилучшего разбиения фичей, в то время как в XGBoost применяется pre-sorted algorithm & Histogram-based algorithm.<br>

LightGBM и Catboost имеют встроенные методы обработки категориальных фичей. Для обучения XGBoost, категориальные переменные необходимо предварительно обработать (например уже реализованными методами из sklearn). При этом Catboost может работать "из коробки" с категориальными фичами в текстовом виде, а для LightGBM необходимо предваритаельно перевести категорильные признаки к типу данных int (например с помощью One Hot Encoding).<br>

LightGBM и Catboost имеют преимущество в скорости работе по сравнению с XGBoost.<br>

Стоит отметить, что при использовании Catboost максимальное качество достигается при использваонии встроенной обработки категорильных фичей с дальнейшей настройкой гиперпараметров.<br>
